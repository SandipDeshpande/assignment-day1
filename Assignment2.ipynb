```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.get_dataset_names()
```

```python
iris=sns.load_dataset('iris')
iris['species'].value_counts()
iris.head()
```

```python
sns.distplot(iris['sepal_width'],color='b')
```

```python
sns.set_style("darkgrid")
sns.distplot(iris['petal_width'],color='g')
```

```python
sns.pairplot(iris,hue='species',size=5)
```

```python
sns.pairplot(iris,hue='petal_length')
```

```python
sns.rugplot(iris['petal_length'],height=1.0,color='r')
```

```python
sns.stripplot(x='species',y='sepal_length',data=iris)
```

```python
sns.stripplot(x='species',y='petal_length',data=iris,hue='sepal_length',dodge=True)
```

```python
sns.violinplot(x='petal_length',y='petal_width',data=iris,hue='species',size=5.0)
```

```python
sns.lmplot(x='petal_length',y='petal_width',data=iris,col='species',size=2.0)
```

```python
sns.FacetGrid(iris,hue='species',size=4)\
    .map(plt.scatter,'petal_length','petal_width')\
    .add_legend();
```

```python
sns.factorplot(x='species',y='sepal_length',data=iris,kind='bar')
```
